```python
# Membuat daftar kondisi untuk digunakan dalam loop
years = [2019, 2022, 2023]

# Menjalankan loop untuk setiap kondisi
for year in years:
    for bulan in range(1, 13):
        # Membuka koneksi
        conn = pyodbc.connect(connection_string)
        # Contoh menjalankan kueri
        query = f""" select 
                store,
                sum(TOTAL_TRX) totaltrx,
                sum(NETSALES) netsls,
                sum(ITMQTY) itmqty,
                count(counter) trxattempt,
                basketsizecategory
                from
                (
            select 
                    1 as counter,
                    year(trxdate) as YR,
                    month(trxdate) as MTH,
                    store as store,
                    NETSALES,
                    ITMQTY,
                    TOTAL_TRX,
                    case
                    when NETSALES <= 100000 then '0-100'
                    when NETSALES > 100000 and NETSALES <= 200000 then '100-200'
                    when NETSALES > 200000 and NETSALES <= 250000 then '200-250'
                    when NETSALES > 250000 and NETSALES <= 275000 then '250-275'
                    when NETSALES > 275000 and NETSALES <= 300000 then '275-300'
                    when NETSALES > 300000 and NETSALES <= 325000 then '300-325'
                    when NETSALES > 325000 and NETSALES <= 350000 then '325-350'
                    when NETSALES > 350000 and NETSALES <= 375000 then '350-375'
                    when NETSALES > 375000 and NETSALES <= 400000 then '375-400'
                    when NETSALES > 400000 and NETSALES <= 450000 then '400-450'
                    when NETSALES > 450000 and NETSALES <= 500000 then '450-500'
                    when NETSALES > 500000 and NETSALES <= 550000 then '500-550'
                    when NETSALES > 550000 and NETSALES <= 600000 then '550-600'
                    when NETSALES > 600000 and NETSALES <=700000 then '600-700'
                    when NETSALES > 700000 and NETSALES <= 800000 then '700-800'
                    when NETSALES > 800000 and NETSALES <= 900000 then '800-900'
                    when NETSALES > 900000 and NETSALES <= 1000000 then '900-1000'
                    when NETSALES > 1000000 and NETSALES <= 1100000 then '1000-1100'
                    when NETSALES > 1100000 and NETSALES <= 1200000 then '1100-1200'
                    when NETSALES > 1200000 and NETSALES <= 1300000 then '1200-1300'
                    when NETSALES > 1300000 and NETSALES <= 1400000 then '1300-1400'
                    when NETSALES > 1400000 and NETSALES <= 1500000 then '1400-1500'
                    when NETSALES > 1500000 and NETSALES <= 1600000 then '1500-1600'
                    when NETSALES > 1600000 and NETSALES <= 1700000 then '1600-1700'
                    when NETSALES > 1700000 and NETSALES <= 1800000 then '1700-1800'
                    when NETSALES > 1800000 and NETSALES <= 1900000 then '1800-1900'
                    when NETSALES > 1900000 and NETSALES <= 2000000 then '1900-2000'
                    when NETSALES > 2000000 then '>2000'
                    else ''
                    end as basketsizecategory -- select top 100 *
                from [STAGING_MDS]..TBL_BASKETSIZE_{year}
               where month(trxdate) = {bulan}
                AND YEAR(TRXDATE) = {year}
            ) a
            group by basketsizecategory, store
            order by store asc
        """

        df = pd.read_sql_query(query, conn)

        # Menampilkan DataFrame
        #print(df)

        # Menutup koneksi
        conn.close()

        #save to excel
        df.to_excel(str(bulan) + '_' + str(year) + '.xlsx', index=False)
```
